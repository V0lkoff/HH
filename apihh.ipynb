```python
import requests
from itertools import product
from functools import reduce
from pandas.io.json import json_normalize
# https://github.com/hhru/api

PARAMS = {
    'experience':['between1And3','noExperience'],
    'specialization':['1.25', '3.26','5.27'],
    'area':'1',
    'per_page':'50'
}

def gen_params_strs(params):
    a=[]
    for key in list(params):
        aa=[]
        for value in params[key]:
            aa.append([key+'='+value])
        a.append(aa)
        
    return list(map(lambda y: '&'.join(reduce(lambda x1,x2: x1+x2, y)), list(product(*a))))

def get_items_by_params(params_strs):
    items=[]
    for param_str in params_strs:
        #query = 'https://api.hh.ru/vacancies?' + param_str + 'text='
        query = 'https://api.hh.ru/vacancies?' + param_str
        resp = requests.get(query, headers = {"User-Agent": "api-test-agent"})
        items+=resp.json()['items']
        
        #print(items)
        
    return items

PARAMS_STRS = gen_params_strs(PARAMS)
#print('params str', PARAMS_STRS)
ITEMS = get_items_by_params(PARAMS_STRS)



```

```python
df = json_normalize(ITEMS[0])
for i in range(1, len(ITEMS)):
    df=df.append(json_normalize(ITEMS[i]))

df
```
